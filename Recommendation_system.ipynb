```python
import ast

def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies_df['genres']=movies_df['genres'].apply(convert)
```

```python
print(movies_df['genres'])
```

```python
movies_df.head(1)
```

```python
movies_df['keywords']=movies_df['keywords'].apply(convert)
```

```python
movies_df.head(1)
```

```python
def convert3(obj):
  M=[]
  counter=0
  for i in ast.literal_eval(obj):
    if(counter!=3):
      M.append(i['name'])
      counter+=1
    else:
      break
  return M


```

```python
movies_df['cast']=movies_df['cast'].apply(convert3)
```

```python
movies_df.head(1)
```

```python
def convert4(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if(i['job']=='Director'):
      L.append(i['name'])
      break
  return L
```

```python
movies_df['crew']=movies_df['crew'].apply(convert4)
```

```python
movies_df.head()
```

```python
movies_df['overview']=movies_df['overview'].apply(lambda x:x.split())
```

```python
movies_df.head()
```

```python
movies_df['genres']=movies_df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies_df['keywords']=movies_df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies_df['cast']=movies_df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies_df['crew']=movies_df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies_df.head()
```

```python
movies_df['tags']=movies_df['overview']+ movies_df['genres']+ movies_df['keywords']+ movies_df['cast']+ movies_df['crew']
```

```python
movies_df.head()
```

```python
new_movies=movies_df[['id','title','tags']]
```

```python
new_movies.head()
```

```python
new_movies['tags']=new_movies['tags'].apply(lambda x: " ".join(x))
```

```python
new_movies['tags']=new_movies['tags'].apply(lambda x:x.lower())
```

```python
new_movies.head()
```

```python
#stemming
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
#function for stemming
def stem (text):
  L=[]
  for i in text.split():
    L.append(ps.stem(i))

  return " ".join(L)
```

```python
new_movies['tags']=new_movies['tags'].apply(stem)
```

```python
new_movies.head()
```

```python
#sklearn module for vectorization and removing stop words
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new_movies['tags']).toarray()
```

```python
print(vector)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
#
def recommend(movie):
  movie_index= new_movies[new_movies['title']==movie].index[0]
  distances=similarity[movie_index]
  movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movie_list:
    print(new_movies.iloc[i[0]].title)
```

```python
recommend('Finding Nemo')
```
